```python
import tensorflow as tf
import matplotlib.pyplot as plt
#tensor:다차원 배열
tf.set_random_seed(777)
X=[1,2,3] 
Y=[1,2,3]
w=tf.placeholder(tf.float32) #의미? 이문장을 w노드로 정의하겠다~할당의 개념이 아닌 정의
#placeholder:실행시 값을 전달받겠다
hf=X*w #b는 생략함, tf포함이어서 변수가 아니라 노드이다
X
```

```python
w
```

```python
hf
```

```python
cost= tf.reduce_mean(tf.square(hf-Y)) #노드  (X*0=[0,0,0])-Y )**2의 각 원소룰 더한 뒤 제곱한 것의 평균은 4.66
cost
```

```python
#다차원배열이 노드사이를 옮겨다닌다.
#그래프(노드(연산)와 텐서(다차원배열)) 정의
sess=tf.Session()
#세션은 서버와 클라이언트가 연결되어 있는 것 세션이 먼저 있어야 그래프 등 실행 가능
w_history=[]
cost_history=[]
for i in range(-30,50):
    curr_w=i*0.1 #-3,-2.9 ~ 4.9
    curr_cost=sess.run(cost, feed_dict={w:curr_w}) #세션객체를 이용하여 cost 노드를 실행하라.
    w_history.append(curr_w)
    cost_history.append(curr_cost)
# print(cost_history)    
plt.plot(w_history, cost_history)
plt.show()
```

```python
tf.set_random_seed(777)
x_train=[1,2,3]
y_train=[1,2,3]
#가설함수:y_train=x_train*w+b에서 최적의 w,b를 찾는다
w=tf.Variable(tf.random_normal([1])) #난수 1개 만들어짐
b=tf.Variable(tf.random_normal([1]))
hf=x_train*w+b
cost=tf.reduce_mean(tf.square(hf-y_train))
optimizer=tf.train.GradientDescentOptimizer(learning_rate=0.01) 
train=optimizer.minimize(cost)

sess=tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(20001):
    sess.run(train)
    if step%100==0:
        print("===",str(step+1),"번 트레이닝 후===")
        print("w=",sess.run(w))
        print("b=",sess.run(b))
        print("예측=",sess.run(hf))
        print("cost=",sess.run(cost))

# sess.run(train)
# print("===1번 트레이닝 수행 후===")
# print("w=",sess.run(w))
# print("b=",sess.run(b))
# print("예측=",sess.run(hf))
# print("cost=",sess.run(cost))

# sess.run(train)
# print("===2번 트레이닝 수행 후===")
# print("w=",sess.run(w))
# print("b=",sess.run(b))
# print("예측=",sess.run(hf))
# print("cost=",sess.run(cost))
```
