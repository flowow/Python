```python
ebola=pd.read_csv("../data/country_timeseries.csv")
print(ebola.columns)
```

```python
ebola_long=pd.melt(ebola, id_vars=['Date','Day'])
ebola_long
```

```python
variable_split=ebola_long.variable.str.split('_')
variable_split[0:4]
```

```python
country_values=variable_split[0][1]
```

```python
ebola_long['status']=status_values
ebola_long['country']=country_values
ebola_long.head()
```

```python
weather=pd.read_csv("../data/weather.csv")
print(weather.iloc[:5,:])
```

```python
weather_melt=pd.melt(weather, id_vars=['id',  'year',  'month', 'element'],
                    var_name='day',value_name='temp')
```

```python
weather_melt
```

```python
weather_melt
```

```python
weather_melt2=weather_melt.pivot_table(index=['id',  'year',  'month', 'day'],
                         columns='element',values='temp')
```

```python
weather_melt2
```

```python
weather_melt2.reset_index()
```

## 중복데이터 처리(중요한 데이터는 살려야)

```python
billboard=pd.read_csv("../data/billboard.csv")
b=pd.melt(billboard, id_vars=['year','artist','track','time'],
       var_name='week',value_name='rating')
```

```python
b
```

```python
b.shape
```

```python
b[b.track=='Loser'] #Loser와 중복데이터 찾기
```

```python
bsong=b[['year','artist','track','time']]
```

```python
bsong.head()
```

```python
b.shape
```

```python
bsong=bsong.drop_duplicates()
bsong
```

```python
bsong['id']=range(len(bsong))
bsong.head(n=10)
```

```python
brating=b.merge(bsong, on=['year','artist','track','time'])
brating.shape
```

```python
brating.head()
```

```python
import os
import urllib.request
with open("../data/raw_data_urls.txt","r") as data_urls: 
    for line,url in enumerate(data_urls): #반복가능한 개체가 들어간다
#         print("line:",line)
#         print("url:",url)
            if line==5:
                break
            fn=url.split("/")[-1].strip()
            print(fn)
            fp=os.path.join('','../data',fn)
            print(fp)
            urllib.request.urlretrieve(url,fp) #url로 접근해서
# 다운로드 파일있는 경로 data_urls
```
