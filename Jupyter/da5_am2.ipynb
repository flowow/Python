```python
def get_counts(sequence):
    counts={}
    for x in sequence:
        if x in counts: #counts 딕셔너리에 x라는 키가 있으면 counts={'x':1}
            counts[x]+=1
        else:
            counts[x]=1
    return counts
get_counts(time_zones)
```

```python
from collections import defaultdict
def get_counts2(sequence):
    counts=defaultdict(int) #counts 딕셔너리 값 0으로 초기화
    for x in sequence:
        counts[x]+=1
    return counts
print(get_counts2(time_zones))
```

```python
counts=get_counts2(time_zones)
```

```python
def top_counts(count_dict, n=10):
    value_key_pairs=[(count,tz) for tz,count #tuple구조가 모인 list
                     in count_dict.items()]
    value_key_pairs.sort()
    print(value_key_pairs[-n:])
    return value_key_pairs[-n:]
top_counts(counts)
```

1) iloc:integer position을 통해 값을 검색, label로는 검색 안 됨
2) loc: label로
3) ix: interger position, label 모두 사용 가능

```python
import pandas as pd
import numpy as np
s=pd.Series(np.nan, index=[49,48,47,46,45,1,2,3,4,5])
print(s)
print("="*50)
print(s.iloc[:3]) #첨부터 3개!(세 번째 인덱스 전까지)
print("="*50)
print(s.loc[:3]) #3까지!
print("="*50)
print(s.ix[:3])#label이 숫자인 경우에는 label 기반 인덱싱 하므로 
#s에대해서는 loc와 동일 결과
```

```python
s2=pd.Series(np.nan, index=['a','b','c','d','e',1,2,3,4,5])
print(s2.iloc[:3])
print("="*50)
# print(s2.loc[:3]) #문자 숫자 섞인 복합형 index는 loc사용 불가
print(s2.ix[:3]) #위와 다르게 iloc->
print("="*50)
```

# 특정 기준으로 그룹화

```python
df=pd.read_csv("data/gapminder.tsv",sep="\t")
print(df)
#year 기준으로 평균구하가
# df.groupby('year').mean('lifeExp')
print(df.groupby('year').lifeExp.mean())
df.year.unique() 
df.loc[df.year==1952,:]
```

```python
def my_mean(values):
    n=len(values)
    sum=0
    for value in values:
        sum+=value
    return sum/n
    print(values)
    
df.groupby('year').lifeExp.agg(my_mean) #년도별 각 데이터 
#lifeEx컬럼을 my_mean함수로 전달
```

```python
def my_mean_diff(values,diff_value):
    n=len(values)
    sum=0
    for value in values:
        sum+=value
    mean=sum/n
    return mean-diff_value
```

## (전체 평균수명 - 해당년도 평균수명) 한번에 출력하기

```python
global_mean=df.lifeExp.mean()
# global_mean
agg_mean_diff= df.groupby('year').lifeExp.agg(my_mean_diff, diff_value=global_mean)
#df를 year 기준으로 그룹화 -> 그룸단위로 lifeExp 값을  my_mean_diff 함수에 전달, 
#이때 diff_value변수에는 global_mean을 전달하라
print(agg_mean_diff)
```

```python
#표준점수(z점수) -> 평균:0, 표준편차:1
#z점수
def my_zscore(x):
    return( (x-x.mean())/x.std())
transorm_z=df.groupby('year').lifeExp.transform(my_zscore)
print(transorm_z)
print(len(df))
df
```

```python
import seaborn as sns
np.random.seed(2019129) #동일한 환경에서의 비교를 위해서
tips_10=sns.load_dataset('tips').sample(10) #sample()은 표본추출
tips_10
```

```python
tips_10.loc[np.random.permutation(tips_10.index)[:4],#tips_10.index에서 4개 임의로 뽑음
            'total_bill']=np.NaN
print(tips_10)
```

```python
count_sex=tips_10.groupby('sex').count()
print(count_sex) #Male의 total_bill data가 많이 손실됨->평균값을 넣든가(KNN이 더 바람직) 
#성별별로 통계치 따로 구해서 넣어주든가 
```

```python
def fill_na_mean(x):
    avg=x.mean()
    return x.fillna(avg)
total_bill_group_mean=tips_10.groupby('sex').total_bill.transform(fill_na_mean)
tips_10['fill_total_bill']=total_bill_group_mean
tips_10
```
