```python
import numpy as np
# np.sum(pd.isnull(titanic['age'])) #True는 1이라서 np.sum하면 nan의 개수 구해짐 
np.sum(pd.isnull(titanic['age']))/titanic['age'].size

def count_missing(vec):
    null_vec=pd.isnull(vec)
    null_count=np.sum(null_vec)
    return null_count
def prop_missing(vec):
    num=count_missing(vec) #Nan의 개수 리턴
    dem=vec.size #전체 데이터 길이
    return num/dem #각 컬럼에 대한 Nan의 비율
def prop_complete(vec):
    return 1-prop_missing(vec)
# cmis_col=titanic.apply(count_missing)
# print(cmis_col)
# pmis_col=titanic.apply(prop_missing)
# print(pmis_col)
rmis_col=titanic.apply(prop_complete)
print(rmis_col)
```

```python
titanic.apply(count_missing)
titanic.apply(count_missing,axis=1).head()
```

```python
titanic.apply(prop_missing)
titanic.apply(prop_missing, axis=1).head()
```

```python
titanic.apply(prop_complete)
```

```python
np.shape(titanic) ######으와 나왔어
```

```python
titanic['num_missing']=titanic.apply(count_missing,axis=1)
```
